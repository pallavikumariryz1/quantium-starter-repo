```python
import pandas as pd
import glob
```

```python
import os
print(os.getcwd())
base_path = os.getcwd()
data_path = os.path.join(base_path, "data", "*.csv")
data_path
```

```python
# step1 : Read all CSV files from data folder
files = glob.glob("*.csv")

# Creating an empty list
df_list = []

for file in files:
    df = pd.read_csv(file)
    df_list.append(df)

# Step 2: Combine all data
data = pd.concat(df_list,ignore_index=True)

# Filter Pink morsels
data = data[data["product"].str.lower()== "pink morsel"]

# Remove $ sign and convert to float
data["price"] = data["price"].replace('[\$,]', '', regex=True).astype(float)

# Now calculate sales
data["sales"] = data["quantity"] * data["price"]

# Filter only required columns
final_data = data[["sales","date","region"]]

# Save output
final_data.to_csv("../output.csv",index=False)
print("Done")
```
